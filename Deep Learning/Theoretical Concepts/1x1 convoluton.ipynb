# What is $1x1$ convolution and why do we need it?

A 1x1 convolution is a type of convolutional layer where the filter size is 1x1. 

One of the biggest advantages of 1x1 convolutions is that they allow you to control the number of channels in your feature maps. For example, suppose you have a feature map with 64 channels, and you want to reduce the number of channels to 32. You can use a 1x1 convolution with 32 filters, and it will effectively reduce the number of channels in the feature map.

Another advantage of 1x1 convolutions is that they are computationally efficient. Since the filter size is so small, they require fewer computations than larger convolutions. This makes them well-suited for use in deeper neural networks, where computational efficiency is essential.
